# Airbnb EDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Load the dataset

data = pd.read_csv('datasets.csv')
data.head()
```

### Data Exploration

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

### Data Cleaning

```python
data.isnull().sum()
```

```python
# dropping all the missing values
data.dropna(inplace=True)
```

```python
 data.duplicated().sum()
```

```python
# deleting all duplicated rows
data.drop_duplicates(inplace=True)
```

```python
# chagning data types
data['id'] = data['id'].astype(object)
data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

### Data Analysis

```python
# Univariate Analysis

# Price Distribution
sns.histplot(data = data , x='price')
```

```python
# identifying outliers

sns.boxplot(data = data, x = 'price')
```

```python
# Filtering Outliers for Price Column
filtered_data = data[data['price']<1500].copy()
sns.boxplot(data = filtered_data , x= "price")
```

```python
sns.histplot(data = filtered_data , x='price')
plt.ylabel('Frequency')
```

### Feature Engineering

```python
# price per bed

filtered_data['price per bed'] = filtered_data['price']/filtered_data['beds']
filtered_data.head()
```

```python
# Analyzing mean price of a bed in different neighbourhood
filtered_data.groupby(by="neighbourhood_group")['price per bed'].mean()
```

### Bivariate Analysis

```python
sns.barplot(data = filtered_data, x ='neighbourhood_group', y = 'price', hue='room_type' )
```

```python
sns.boxplot(data=filtered_data, x='room_type', y='price')
plt.ylim(0, 1000)
plt.title('Room Type vs Price')
```

```python
# Relationship between number of reviews and price grouped by different neighbourhood

plt.figure(figsize=(6,6))
sns.scatterplot(data = filtered_data, x = 'number_of_reviews', y = 'price', hue = 'neighbourhood_group')
```

```python
sns.pairplot(data = filtered_data, vars = ['price','minimum_nights','number_of_reviews','availability_365'], hue = 'room_type')
```

```python
# heat map
corr = filtered_data[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr
```

```python
sns.heatmap(data = corr, annot = True)
```

```python
filtered_data['rating'] = pd.to_numeric(filtered_data['rating'], errors='coerce')
filtered_data['rating_bin'] = pd.cut(filtered_data['rating'], bins=5)
sns.boxplot(data=filtered_data, x='rating_bin', y='price')
plt.title('Price by Rating Range')
```
